```python
import yfinance as yf
import datetime as dt
import pandas as pd
```

```python
data_inicio = dt.date(2015, 1, 1)
data_final = dt.date(2025, 8, 14)
```

```python
path = 'https://raw.githubusercontent.com/GunzLauncher/portfolios_investment_analysis/refs/heads/main/acoes-listadas-b3.csv'
ativos = pd.read_csv(path, sep=',')
todos_ativos = [f'{s}.SA' for s in ativos['Ticker']]
selecao_ativos = ['WEGE3.SA', 'PETR4.SA', 'VALE3.SA']
```

```python

precos_ativos = yf.download(todos_ativos, data_inicio, data_final)
colunas_remover = ['Close', 'High', 'Low', 'Volume']
precos_ativos.drop(columns=colunas_remover, inplace=True)
precos_ativos
```
