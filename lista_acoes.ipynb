```python
data_inicio = dt.date(2020, 1, 1)
data_final = dt.date(2025, 8, 14)
```

```python
path = 'https://raw.githubusercontent.com/GunzLauncher/portfolios_investment_analysis/refs/heads/main/acoes-listadas-b3.csv?token=GHSAT0AAAAAADIX7Z6DYYYATUJOLCTYRNNO2E6XBSA'
ativos = pd.read_csv(path, sep=',')
todos_ativos = [f'{s}.SA' for s in ativos['Ticker']]
```

```python

df_ativos = yf.download(todos_ativos, data_inicio, data_final)
colunas_remover = ['Open', 'High', 'Low', 'Volume', 'Adj Close']
df_ativos.drop(columns=colunas_remover, inplace=True)
precos_fechamento_ativos = df_ativos.droplevel(0, axis=1)
precos_fechamento_ativos.head()
```
